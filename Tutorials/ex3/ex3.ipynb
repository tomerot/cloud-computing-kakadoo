```python
url = 'https://data.gov.il/api/3/action/datastore_search?resource_id=053cea08-09bc-40ec-8f7a-156f0677aff3'

import requests
import pandas as pd
pd.set_option("display.max_columns", None)

response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)

# Create a DataFrame from the 'records' in the JSON response
data_df = pd.DataFrame(data['result']['records'])
```

```python
from ipywidgets import widgets, Output
from IPython.display import display

output_area = Output() # Anything written to output widget will appear in the Jupyter notebook
```

```python
# Create dropdown widgets to select Tozeret and Kinuy Mishari

tozeret_nm_dropdown = widgets.Dropdown(
    options = [''] + data_df['tozeret_nm'].unique().tolist(),
)

kinuy_mishari_dropdown = widgets.Dropdown(
    options = [''],
)
```

```python
# Function to update the kinuy mishari dropdown
def update_kinuim(change):
    if change['new']:  # Check if a valid option is selected
        kinuim = data_df[data_df['tozeret_nm'] == change['new']]['kinuy_mishari'].unique()
        kinuy_mishari_dropdown.options = [''] + list(kinuim)
    else:
        kinuy_mishari_dropdown.options = ['']
```

```python
# Function to update the output area after kinuy mishari selected
def update_output(change):
    output_area.clear_output() # Clear last output

    # Assign selected tozeret and kinuy mishari to variables
    tozeret = tozeret_nm_dropdown.value
    kinuy = kinuy_mishari_dropdown.value

    if tozeret and kinuy: # Do this only when both were selected, so empty output will not be displayed
      df_selected = data_df[(data_df['tozeret_nm'] == tozeret) & (data_df['kinuy_mishari'] == kinuy)] # Filter data
      total_records = len(df_selected)
      ramot_gimur = df_selected['ramat_gimur'].unique()

      # Output prints
      with output_area:
          print(f"Total Records: {total_records}")
          print(f"Unique Ramat Gimur: {', '.join(ramot_gimur)}")
```

```python
# Set the event listeners for the dropdowns
tozeret_nm_dropdown.observe(update_kinuim, names = 'value')
kinuy_mishari_dropdown.observe(update_output, names = 'value')
```

```python
# Organize the widgets in a structured grid
form_layout = widgets.GridBox(
    children = [
        widgets.Label("Tozeret:"), tozeret_nm_dropdown,
        widgets.Label("Kinuy Mishari:"), kinuy_mishari_dropdown,
    ],
    layout=widgets.Layout(
        grid_template_columns = "100px auto",  # Two columns: label (width of 100px) and field (automatically sized)
        grid_gap = "5px",  # Space between elements
    )
)
```

```python
# Creating space between form_layout and output_area
spacer = widgets.Box(layout = widgets.Layout(height = "10px"))
```

```python
display(form_layout)
display(spacer)
display(output_area)
```
